### Date Format Conversion

Converted Order Date and Ship Date columns from object datatype 
to datetime format for time-series analysis.

The dataset follows DD/MM/YYYY format.

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    dayfirst=True
    )

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    dayfirst=True
    )
```

```python
df["Year"] = df["Order Date"].dt.year

df["Month"] = df["Order Date"].dt.month
```
